# Create a file from two datasets and add headers.

```python
import pandas as pd

r_cols = ['movie_id','user_id' , 'rating']
ratings = pd.read_csv('C:\\Users\\Diego Alves\\Desktop\\test_jupyter.csv', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('C:\\Users\\Diego Alves\Desktop\\test_jupyter_2.csv', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings['WHITE METAL LANTERN']
starWarsRatings.head()
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 3
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
```
